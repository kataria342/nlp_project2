```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
import numpy as np
```

```python
symptomsDF = pd.read_csv('dataset.csv')
```

```python
#symptomsDF.info()
```

```python
symptomsDF = symptomsDF.replace(np.NaN, 'None')

cols = list(symptomsDF.columns)
cols
for i in cols[1:]:
  symptomsDF[i] = symptomsDF[i].str.split("_")

symptomsDF.head(2)
```

```python
symptomsDF['combined_symptoms'] = symptomsDF['Symptom_1'] + symptomsDF['Symptom_2'] + symptomsDF['Symptom_3'] + symptomsDF['Symptom_4'] + symptomsDF['Symptom_5'] + symptomsDF['Symptom_6'] + symptomsDF['Symptom_7'] + symptomsDF['Symptom_8'] + symptomsDF['Symptom_9'] + symptomsDF['Symptom_10'] + symptomsDF['Symptom_11'] + symptomsDF['Symptom_12'] + symptomsDF['Symptom_13'] + symptomsDF['Symptom_14'] + symptomsDF['Symptom_15'] + symptomsDF['Symptom_16'] + symptomsDF['Symptom_17']
symptomsDF.drop(columns=cols[1:],inplace=True)
symptomsDF.head(2)
```

```python
testDF = symptomsDF.copy()
testDF['combined_symptoms'] = testDF['combined_symptoms'].apply(lambda x: ' '.join(x))
testDF.head(2)
```

```python
def clean_duplicate(text):
  words = text.split()
  unique_words = set()
  result = []
  for word in words:
      if word not in unique_words:
          unique_words.add(word)
          result.append(word)

  return ' '.join(result)

testDF['combined_symptoms'] = testDF['combined_symptoms'].apply(clean_duplicate)
testDF.head(2)
```

```python
testDF.shape
```

```python
testDF = testDF.drop_duplicates()
testDF.shape
```

```python
testDF['Disease'].unique()
```

```python
diseaseDescriptionDF = pd.read_csv('symptom_Description.csv')
```

```python
diseaseDescriptionDF.info()
```

```python
diseasePrecautionDF = pd.read_csv('symptom_precaution.csv')
```

```python
diseaseDescriptionDF.info()
```

```python
merged_df = pd.merge(diseaseDescriptionDF, diseasePrecautionDF, on='Disease', how = 'left')
```

```python
merged_df.info()
```

```python
merged_df.set_index('Disease',inplace=True)
merged_df.head(2)
```

```python
merged_df = merged_df.replace(np.NaN, 'None')
```

```python
import tensorflow as tf
import tensorflow_hub as hub
module_url = "https://tfhub.dev/google/universal-sentence-encoder/4"
model = hub.load(module_url)
embed = model(testDF['combined_symptoms'])
```

```python
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors=1)

neigh.fit(embed)
```

```python
from textblob import TextBlob
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def cleandata(text):
    tag_mapping = {
    'NN': 'n',
    'NNS': 'n',
    'NNP': 'n',
    'NNPS': 'n',
    'VB': 'v',
    'VBD': 'v',
    'VBG': 'v',
    'VBN': 'v',
    'VBP': 'v',
    'VBZ': 'v',
    'JJ': 'a',
    'JJR': 'a',
    'JJS': 'a',
    'RB': 'r',
    'RBR': 'r',
    'RBS': 'r',
    'WRB':'r'
}
    text = "".join([word.lower() for word in re.sub(r"[^a-zA-Z]", " ", text)])
    tokens = word_tokenize(text)
    text = [word for word in tokens if len(word)>2 or word not in stop_words]
    text = " ".join([lemmatizer.lemmatize(word, pos='n' if tag_mapping.get(TextBlob(word).tags[0][1]) == None else tag_mapping.get(TextBlob(word).tags[0][1])) for word in text])
    return text
```

```python
symptoms = input('enter your symptoms')
symptoms = cleandata(symptoms)
print(symptoms)
emd = model([symptoms])
nei = neigh.kneighbors(emd,return_distance=False)[0][0]
disease_name = testDF['Disease'].iloc[nei]
description = merged_df['Description'].loc[disease_name]
Precauions = list(merged_df[['Precaution_1','Precaution_2','Precaution_3','Precaution_4']].loc[disease_name])
print(disease_name)
print('Description : ', description)
print('Precautions:')
for i in Precauions:
  print(i)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity_matrix = cosine_similarity(embed, emd)
```

```python
max = cosine_similarity_matrix[0]
index = 0
for i in range(1,len(cosine_similarity_matrix)):
    if cosine_similarity_matrix[i] > max:
        max = cosine_similarity_matrix[i]
        index = i
print(f'Index of the maximum value is : {index}')
```

```python
disease_name = testDF['Disease'].iloc[index]
description = merged_df['Description'].loc[disease_name]
Precauions = list(merged_df[['Precaution_1','Precaution_2','Precaution_3','Precaution_4']].loc[disease_name])
print(disease_name)
print('Description : ', description)
print('Precautions:')
for i in Precauions:
  print(i)
```
